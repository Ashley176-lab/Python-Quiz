```python
def python_quiz():
    score = 0
    answer1 = input("What keyword is used to define a function in Python? ")
    if answer1.lower() == "def":
        score += 1
        
    answer2 = input("What symbol is used for comments in Python? ")
    if answer2 == "#":
        score += 1

    answer3 = input("What data type is True or False called? ")
    if answer3.lower() == "boolean":
        score += 1

    answer4 = input("What function shows output in Python? ")
    if answer4.lower() == "print":
        score += 1

    answer5 = input("What symbol is used for power in Python? ")
    if answer5 == "**":
        score += 1

    print("You got " + str(score) + "/ 5 correct!!")
    continue_quiz()


def it_quiz():
    score = 0
    answer1 = input("What does CPU stand for? ")
    if answer1.lower() == "central processing unit":
        score += 1
    
    answer2 = input("What does RAM stand for? ")
    if answer2.lower() == "random access memory":
        score += 1

    answer3 = input("What is the main branch of the computer? ")
    if answer3.lower() == "cpu":
        score += 1

    answer4 = input("What device connects a computer to the internet? ")
    if answer4.lower() == "router":
        score += 1

    answer5 = input("What do we call programs on a computer? ")
    if answer5.lower() == "software":
        score += 1

    print("You got " + str(score) + "/ 5 correct!!")
    continue_quiz()


def networking_quiz():
    score = 0
    answer1 = input("What does IP stand for? ")
    if answer1.lower() == "internet protocol":
        score += 1

    answer2 = input("What does HTTP stand for? ")
    if answer2.lower() == "hypertext transfer protocol":
        score += 1

    answer3 = input("What protects a network from attacks? ")
    if answer3.lower() == "firewall":
        score += 1

    print("You got " + str(score) + "/ 3 correct!!")
    continue_quiz()
    

def continue_quiz():
    option2 = input("Would you like to continue with other quizzes? ")
    if option2.lower() == "no":
        print("Thank you for participating!! Have a great day!!")
        quit()
    else:
        print("Great!What knowledge would you like to test from the list below? ")
        print("Python")
        print("IT")
        print("Networking")
        option = input("Please enter your preference: ")
        if option.lower() == "python":
            python_quiz()

        if option.lower() == "it":
            it_quiz()

        if option.lower() == "networking":
            networking_quiz()
        
        
def my_fun_quiz():
    print("Welcome To The Fun IT Quiz! ")

    play = input("Wanna play? **in Chucky's voice** : ")
    if play.lower() == "no" :
        quit()
    else:
        print("Great!What knowledge would you like to test from the list below? ")
        print("Python")
        print("IT")
        print("Networking")
        option = input("Please enter your preference: ")
        
        if option.lower() == "python":
            python_quiz()

        if option.lower() == "it":
            it_quiz()

        if option.lower() == "networking":
            networking_quiz()

my_fun_quiz()
```
